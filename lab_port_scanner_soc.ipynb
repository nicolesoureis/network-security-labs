```python
import socket

alvo = "scanme.nmap.org"

portas = [21, 22, 80, 443]

for porta in portas:

    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(1)

    resultado = s.connect_ex((alvo, porta))

    if resultado == 0:
        print("Porta aberta:", porta)

    else:
        print("Porta fechada:", porta)

    s.close()
```

# LAB 01 — Scanner de Portas TCP com Python

Este laboratório demonstra a criação de um scanner de portas utilizando Python e a biblioteca socket.

## Objetivo

Identificar portas abertas em um host remoto para mapear serviços expostos na rede.

## Tecnologias utilizadas

- Python
- Socket
- Google Colab

## Portas analisadas

21 — FTP  
22 — SSH  
80 — HTTP  
443 — HTTPS  

## Resultado

O script identifica quais portas estão abertas e quais estão fechadas, simula uma etapa inicial de reconnaissance de rede utilizada em atividades de análise de superfície de ataque em ambientes SOC.
